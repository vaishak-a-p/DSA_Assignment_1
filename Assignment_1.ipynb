### 5. Visualizations

Scatter plot of SW vs SL:

```python
from matplotlib import pyplot as plt
plt.scatter(x = iris_data['SL'], y = iris_data['SW'])
plt.title('SW vs SL Graph', fontdict={'fontsize': 14})
plt.xlabel('SL')
plt.ylabel('SW')
plt.show()
```

Scatter Plot of PW vs PL:

```python
plt.scatter(x = iris_data['PL'], y = iris_data['PW'], c='orange')
plt.title('PW vs PL Graph', fontdict={'fontsize': 14})
plt.xlabel('PL')
plt.ylabel('PW')
plt.show()
```

Pair Plot of SW, SL, PW and PL:

```python
import seaborn as sns
sns.pairplot(iris_data, hue='Classification')
plt.show()
```

Heatmap of Pairwise Correlation Between SL, SW, PL and PW:

```python
sns.heatmap(iris_data.corr(), linecolor='white', linewidths=2, annot=True)
plt.show()
```
